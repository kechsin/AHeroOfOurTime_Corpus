```python
import requests
from bs4 import BeautifulSoup
import csv

# URL страницы словаря
url = "https://azbyka.ru/otechnik/Spravochniki/slovar-ustarevshih-slov/#0_1"

# Получаем HTML-код страницы
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Находим раздел с устаревшими словами
words_section = soup.find_all("p", class_="txt")

old_words = []
for word_entry in words_section[14:]:  # почистили информацию и оставили только слова
    old_word_line = word_entry.text.strip().split(' – ')
    old_words += [old_word_line[0].lower()]
old_words_final = []
for word in old_words:
  word = word.split(', ')[0]
  word = word.split(' (')[0]
  word = word.split(' ')[0]
  old_words_final += [word]
print(old_words_final)
```

```python
with open('old_words.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Слово'])  # Заголовок для колонки

    # Записываем каждое слово в отдельной строке
    for word in old_words_final:
        writer.writerow([word])
```

```python
import pandas as pd

# Загрузка датасета
df_corpus = pd.read_csv("corpus_data.csv")
df_old_words = pd.read_csv("old_words.csv")

# Вывод третьего столбца
lemmas = df_corpus.iloc[:, 2].fillna('')
labels = []
old_words_list = df_old_words['Слово'].tolist()

for word_list in lemmas:
    labels_of_one_sentence = []
    words = word_list.split(';')  # Разбиваем строку на слова по разделителю ";"
    for word in words:
        if word in old_words_list:
            labels_of_one_sentence.append(1)
        else:
            labels_of_one_sentence.append(0)
    labels.append(labels_of_one_sentence)

print(labels)
```

```python
labels_as_strings = [';'.join(map(str, label_list)) for label_list in labels]

# Добавляем новый столбец в DataFrame
df_corpus['Is_Obsolete'] = labels_as_strings

# Сохраняем обновленный DataFrame
df_corpus.to_csv("corpus_data_with_labels.csv", index=False)
```
